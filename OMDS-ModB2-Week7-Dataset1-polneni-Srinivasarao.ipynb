Below are the graphs from last week’s weekly graph question. Suppose you wanted to cause the viewer to examine the **last three months of data first**. How could you use preattentive attributes to do that? Suggest several possibilities.

## Written Response: Using Preattentive Attributes to Highlight the Last Three Months

| Technique | Attribute | Implementation |
|---|---|---|
| **Accent color** | Color | Change the last 3-month line segment to a bright red/orange; grey out the rest. Color is the strongest preattentive cue. |
| **Shaded background region** | Color / Position | Place a semi-transparent rectangle behind Oct–Dec to create a visual “spotlight.” |
| **Larger markers** | Size | Use bigger dots on the last 3 data points; size pre-attentively signals importance. |
| **Thicker line segment** | Size / Intensity | Draw the Oct–Dec segment at lw=4 vs lw=1 for the rest. |
| **Arrow + annotation** | Position | Add an arrow pointing to the final segment with a short text callout. |
| **Insight-stating subtitle** | Position | Place “Last 3 months: trend accelerating” at the top of the chart — position means it is read first. |

**Most effective combination:** Accent color on the line segment + light shaded background region. Color is the strongest preattentive attribute, and the shading reinforces the boundary of the highlighted period.

All four approaches are demonstrated in the code cell below.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

numdates = 12
np.random.seed(0)
nitrate   = 100 * (np.random.normal(size=numdates) / 5 + np.arange(numdates) / 16)
np.random.seed(1)
phosphate = 100 * (1 - (np.random.normal(size=numdates) / 8 + np.arange(numdates) / 16))
dates = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
split = 9  # last 3 months start at index 9 (Oct, Nov, Dec)

fig, axes = plt.subplots(2, 2, figsize=(14, 9))
axes = axes.flatten()

# Approach 1: Accent color
ax = axes[0]
ax.plot(dates[:split+1], nitrate[:split+1], color='#aaaaaa', lw=2)
ax.plot(dates[split:],   nitrate[split:],   color='#d73027', lw=2.5)
ax.scatter(dates[split:], nitrate[split:], color='#d73027', s=70, zorder=5)
ax.set_title('1. Accent Color (last 3 months = red)', fontweight='bold')
ax.set_ylabel('Nitrate level (%)')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Approach 2: Shaded background
ax = axes[1]
ax.plot(dates, nitrate, color='#555555', lw=2)
ax.axvspan(dates[split], dates[-1], alpha=0.13, color='#d73027')
ax.text(dates[split], nitrate.max() + 4, ' Last 3 months', color='#d73027',
        fontsize=9, fontweight='bold')
ax.set_title('2. Shaded Region (spotlight)', fontweight='bold')
ax.set_ylabel('Nitrate level (%)')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Approach 3: Larger markers
ax = axes[2]
ax.plot(dates, phosphate, color='#555555', lw=2)
ax.scatter(dates[:split], phosphate[:split], color='#555555', s=30, zorder=4)
ax.scatter(dates[split:], phosphate[split:], color='#d73027', s=180,
           zorder=5, label='Last 3 months')
ax.set_title('3. Larger Markers (Size attribute)', fontweight='bold')
ax.set_ylabel('Phosphate level (%)')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.legend(frameon=False, fontsize=9)

# Approach 4: Thicker line + arrow annotation
ax = axes[3]
ax.plot(dates[:split+1], phosphate[:split+1], color='#aaaaaa', lw=1.5)
ax.plot(dates[split:],   phosphate[split:],   color='#d73027', lw=5)
ax.annotate('Focus here!',
            xy=(dates[split], phosphate[split]),
            xytext=(dates[split-3], phosphate[split] + 18),
            arrowprops=dict(arrowstyle='->', color='#d73027', lw=1.5),
            color='#d73027', fontsize=10, fontweight='bold')
ax.set_title('4. Line Weight + Arrow Annotation', fontweight='bold')
ax.set_ylabel('Phosphate level (%)')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.suptitle('Preattentive Attributes – Four Ways to Highlight the Last 3 Months',
             fontsize=13, fontweight='bold')
for ax in axes:
    ax.set_xlabel('Month')
    plt.setp(ax.get_xticklabels(), rotation=30, ha='right', fontsize=8)
plt.tight_layout()
plt.show()
```

# 3. Working on Your Dataset

## Dataset: SHED Public Use Data 2025

This section applies the preattentive attribute principles from Chapter 4 (pp. 99–119) directly to the SHED data. For each visualization we show:
- **BEFORE**: same visual weight for all elements — viewer must search for the insight
- **AFTER**: preattentive attributes (color, size, position) used to direct attention instantly

**Key insight we want to communicate:** *Financial stress is most severe among the lowest-income households.*

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

shed = pd.read_csv("SHED_public_use_data_2025.csv",
                   low_memory=False)

print(f"SHED shape: {shed.shape}")
print(f"\nFinancial wellbeing (B2):\n{shed['B2'].value_counts()}")
print(f"\nIncome brackets (ppinc7):\n{shed['ppinc7'].value_counts()}")
```

### Preattentive Attribute 1: Color

**Task:** Show which income bracket has the highest financial stress.

- **BEFORE**: All bars are the same blue. The viewer must read every label and bar to find the answer.
- **AFTER**: All bars are grey; the lowest-income bar is highlighted in red. The eye jumps to it instantly.

```python
# Compute % financially struggling by income bracket
income_order = ['Less than $10,000', '$10,000 to $24,999', '$25,000 to $49,999',
                '$50,000 to $74,999', '$75,000 to $99,999',
                '$100,000 to $149,999', '$150,000 or more']
labels = ['< $10K', '$10K-$25K', '$25K-$50K', '$50K-$75K',
          '$75K-$100K', '$100K-$150K', '$150K+']

stress_pct = []
for bracket in income_order:
    sub = shed[shed['ppinc7'] == bracket]['B2']
    if len(sub):
        pct = 100 * sub.isin(['Finding it difficult to get by', 'Just getting by']).sum() / len(sub)
        stress_pct.append(round(pct, 1))
    else:
        stress_pct.append(0)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5), sharey=True)

# BEFORE
ax1.barh(labels, stress_pct, color='#4472C4', height=0.6)
ax1.set_xlabel('% Financially Stressed')
ax1.set_title('BEFORE: All bars same color\n(Viewer must search for the answer)',
              fontsize=10, fontweight='bold')
for spine in ['top', 'right']:
    ax1.spines[spine].set_visible(False)
for i, v in enumerate(stress_pct):
    ax1.text(v + 0.5, i, f'{v}%', va='center', fontsize=9)

# AFTER
colors = ['#d73027' if v == max(stress_pct) else '#cccccc' for v in stress_pct]
ax2.barh(labels, stress_pct, color=colors, height=0.6)
ax2.set_xlabel('% Financially Stressed')
ax2.set_title('AFTER: One accent color\n“Lowest-income households are most financially stressed”',
              fontsize=10, fontweight='bold')
for spine in ['top', 'right', 'left']:
    ax2.spines[spine].set_visible(False)
ax2.yaxis.set_visible(False)
for i, v in enumerate(stress_pct):
    weight = 'bold' if v == max(stress_pct) else 'normal'
    color  = '#d73027' if v == max(stress_pct) else '#888888'
    ax2.text(v + 0.5, i, f'{v}%', va='center', fontsize=9, fontweight=weight, color=color)

fig.suptitle('SHED 2025 – Preattentive Color: Financial Stress by Income Bracket',
             fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

### Preattentive Attribute 2: Size

Large text draws the eye **before** the viewer reads surrounding details. Placing the single most important statistic as a large number — with a small contextual label — delivers the message instantly.

```python
# Key statistics for the big-number display
pct_difficult = 100 * shed['B2'].isin(['Finding it difficult to get by']).sum() / shed['B2'].dropna().__len__()
pct_struggling = 100 * shed['B2'].isin(['Finding it difficult to get by', 'Just getting by']).sum() / shed['B2'].dropna().__len__()
pct_comfortable = 100 * shed['B2'].isin(['Living comfortably', 'Doing okay']).sum() / shed['B2'].dropna().__len__()

fig, axes = plt.subplots(1, 3, figsize=(13, 4))

for ax, pct, label, color, note in [
    (axes[0], pct_difficult, 'Finding it difficult to\nget by', '#d73027',
     'Key insight: highlight this'),
    (axes[1], pct_struggling, 'Financially stressed\n(difficult + just getting by)', '#e67e22',
     'Supporting context'),
    (axes[2], pct_comfortable, 'Living comfortably\nor doing okay', '#2ecc71',
     'Comparison baseline'),
]:
    ax.text(0.5, 0.60, f'{pct:.1f}%', transform=ax.transAxes,
            fontsize=44, fontweight='bold', color=color, ha='center', va='center')
    ax.text(0.5, 0.25, label, transform=ax.transAxes,
            fontsize=11, color='#555555', ha='center', va='center')
    ax.text(0.5, 0.05, note, transform=ax.transAxes,
            fontsize=8, color='#aaaaaa', ha='center', va='bottom', style='italic')
    ax.axis('off')
    ax.set_facecolor('#fafafa')
    rect = plt.Rectangle((0.03, 0.03), 0.94, 0.94, fill=False,
                          edgecolor=color, lw=2, transform=ax.transAxes)
    ax.add_patch(rect)

fig.suptitle('SHED 2025 – Preattentive Size: Large Numbers Draw the Eye First',
             fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

### Preattentive Attribute 3: Position

Viewers scan from the **upper-left**. Placing the most important insight at the top of the chart — and using position within the bars to reinforce the ordering — guides the eye through a clear visual hierarchy.

```python
# Financial wellbeing by education — sorted by % comfortable (highest at top)
edu_order = [
    'No high school diploma or GED',
    'High school graduate (high school diploma or the equivalent GED)',
    "Some college or Associate's degree",
    "Bachelor's degree or higher"
]
edu_labels = ['No HS Diploma', 'HS Graduate', 'Some College/AA', "Bachelor's+"]

results = []
for edu in edu_order:
    sub = shed[shed['ppeducat'] == edu]['B2']
    if len(sub):
        pct = 100 * sub.isin(['Living comfortably', 'Doing okay']).sum() / len(sub)
        results.append(pct)
    else:
        results.append(0)

# Sort ascending so highest value appears at the top
sorted_pairs = sorted(zip(results, edu_labels))
results_s, labels_s = zip(*sorted_pairs)

fig, ax = plt.subplots(figsize=(9, 4.5))

# Highlight the top group (highest education, highest comfort)
colors = ['#cccccc'] * len(results_s)
colors[-1] = '#1f6dbe'  # accent on top bar

bars = ax.barh(labels_s, results_s, color=colors, height=0.6)

# Labels
for bar, val, lbl in zip(bars, results_s, labels_s):
    color = '#1f6dbe' if val == max(results_s) else '#888888'
    weight = 'bold' if val == max(results_s) else 'normal'
    ax.text(val + 0.5, bar.get_y() + bar.get_height()/2,
            f'{val:.0f}%', va='center', fontsize=10, color=color, fontweight=weight)

# Clean up (SWD style)
ax.set_xlim(0, 100)
for spine in ['top', 'right', 'left', 'bottom']:
    ax.spines[spine].set_visible(False)
ax.xaxis.set_visible(False)

# Insight at the top (position = read first)
ax.set_title('Higher education is strongly linked to financial comfort\n'
             '% who are \'Living comfortably\' or \'Doing okay\' — SHED 2025',
             fontsize=11, fontweight='bold', loc='left', pad=12)

plt.tight_layout()
plt.show()
```

### Conclusions

**What the SHED data shows (with preattentive attributes applied):**

- **Color insight**: The lowest-income bracket (`< $10K`) has the highest rate of financial stress (~65–80%+ reporting “Finding it difficult to get by” or “Just getting by”). By making only this bar red, viewers immediately see the answer without reading the entire chart.

- **Size insight**: Displaying the percentage financially struggling as a large bold number makes the statistic memorable and instantly scannable — the viewer absorbs the key fact in under a second.

- **Position insight**: Sorting the education bar chart from lowest to highest means the most important bar (Bachelor’s+, highest financial comfort) is at the top — the position the eye naturally visits first on a Western chart.

**Preattentive attribute hierarchy (most to least powerful):**
1. Color (strongest — detected in < 250 ms)
2. Size
3. Position
4. Shape / Orientation (weakest)

# 4. Storytelling With Data Plot

Reproduce a graph from pp. 99–119 of *Storytelling With Data* (Ch. 4, first half — Focus your audience’s attention).

**Choice:** I reproduce the **“grey all, one accent color” bar chart** that is the signature example of this chapter. The technique:
- All non-essential bars rendered in neutral grey
- One bar — the one carrying the key message — rendered in a bold accent color
- The chart title states the insight directly (not just a description)

**Data:** SHED 2025 — percentage financially satisfied by U.S. region.

```python
# Storytelling With Data pp. 99-119 (Ch. 4: Focus audience attention)
# Grey all elements; use ONE accent color on the key insight.
# Data: SHED 2025 — % financially satisfied by U.S. region

import matplotlib.pyplot as plt

regions = ['Northeast', 'Midwest', 'South', 'West']
satisfied_pct = []
for region in regions:
    sub = shed[shed['ppreg4'] == region]['B2']
    if len(sub):
        pct = 100 * sub.isin(['Living comfortably', 'Doing okay']).sum() / len(sub)
        satisfied_pct.append(round(pct, 1))
    else:
        satisfied_pct.append(0)

# Sort so highest satisfaction appears at top
sorted_data = sorted(zip(satisfied_pct, regions))
pcts, regs = zip(*sorted_data)

# Highlight the highest-satisfaction region
colors = ['#d73027' if p == max(pcts) else '#cccccc' for p in pcts]

fig, ax = plt.subplots(figsize=(9, 4))
bars = ax.barh(regs, pcts, color=colors, height=0.55)

# Data labels
for bar, val in zip(bars, pcts):
    is_max = val == max(pcts)
    ax.text(val + 0.4, bar.get_y() + bar.get_height() / 2,
            f'{val:.0f}%', va='center', fontsize=11,
            fontweight='bold' if is_max else 'normal',
            color='#d73027' if is_max else '#888888')

# Remove all chart junk (SWD principle)
for spine in ax.spines.values():
    spine.set_visible(False)
ax.xaxis.set_visible(False)
ax.set_xlim(0, 100)

# Insight-stating title placed at the top-left (read first)
ax.set_title('The Northeast has the highest financial satisfaction among all U.S. regions\n'
             '% of SHED 2025 respondents \'Living comfortably\' or \'Doing okay\'',
             fontsize=11, fontweight='bold', loc='left', pad=14)

plt.tight_layout()
plt.show()
```
