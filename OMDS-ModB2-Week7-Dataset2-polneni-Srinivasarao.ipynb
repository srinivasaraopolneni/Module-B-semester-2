# 2. Weekly Graph Question

Below are the graphs from last week’s weekly graph question. Suppose you wanted to cause the viewer to examine the **last three months of data first**. How could you use preattentive attributes to do that? Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size=numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x=datearray, y=time_series)
ax.set_xlabel('Year and month')
ax.set_ylabel('Nitrate level (% of eutrophic value)')
plt.title('Nitrate Level Over Time')
plt.tight_layout()
plt.show()
```

## Written Response: Using Preattentive Attributes to Highlight the Last Three Months

| Technique | Attribute | Implementation |
|---|---|---|
| **Accent color** | Color | Change the last 3-month line segment to a bright red/orange; grey out the rest. Color is the strongest preattentive cue. |
| **Shaded background region** | Color / Position | Place a semi-transparent rectangle behind Oct–Dec to create a visual “spotlight.” |
| **Larger markers** | Size | Use bigger dots on the last 3 data points; size pre-attentively signals importance. |
| **Thicker line segment** | Size / Intensity | Draw the Oct–Dec segment at lw=4 vs lw=1 for the rest. |
| **Arrow + annotation** | Position | Add an arrow pointing to the final segment with a short text callout. |
| **Insight-stating subtitle** | Position | Place “Last 3 months: trend accelerating” at the top — position ensures it is read first. |

**Most effective combination:** Accent color on the line segment + light shaded background region. Color is the strongest preattentive attribute, and the shading reinforces the boundary.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

numdates = 12
np.random.seed(0)
nitrate   = 100 * (np.random.normal(size=numdates) / 5 + np.arange(numdates) / 16)
np.random.seed(1)
phosphate = 100 * (1 - (np.random.normal(size=numdates) / 8 + np.arange(numdates) / 16))
dates = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
split = 9

fig, axes = plt.subplots(2, 2, figsize=(14, 9))
axes = axes.flatten()

for ax, series, title, approach in [
    (axes[0], nitrate, '1. Accent Color (last 3 months = red)', 'color'),
    (axes[1], nitrate, '2. Shaded Region (spotlight)', 'shade'),
    (axes[2], phosphate, '3. Larger Markers (Size attribute)', 'size'),
    (axes[3], phosphate, '4. Line Weight + Arrow Annotation', 'weight'),
]:
    if approach == 'color':
        ax.plot(dates[:split+1], series[:split+1], color='#aaaaaa', lw=2)
        ax.plot(dates[split:], series[split:], color='#d73027', lw=2.5)
        ax.scatter(dates[split:], series[split:], color='#d73027', s=70, zorder=5)
    elif approach == 'shade':
        ax.plot(dates, series, color='#555555', lw=2)
        ax.axvspan(dates[split], dates[-1], alpha=0.13, color='#d73027')
        ax.text(dates[split], series.max() + 4, ' Last 3 months',
                color='#d73027', fontsize=9, fontweight='bold')
    elif approach == 'size':
        ax.plot(dates, series, color='#555555', lw=2)
        ax.scatter(dates[:split], series[:split], color='#555555', s=30, zorder=4)
        ax.scatter(dates[split:], series[split:], color='#d73027', s=180,
                   zorder=5, label='Last 3 months')
        ax.legend(frameon=False, fontsize=9)
    elif approach == 'weight':
        ax.plot(dates[:split+1], series[:split+1], color='#aaaaaa', lw=1.5)
        ax.plot(dates[split:], series[split:], color='#d73027', lw=5)
        ax.annotate('Focus here!',
                    xy=(dates[split], series[split]),
                    xytext=(dates[split-3], series[split] + 18),
                    arrowprops=dict(arrowstyle='->', color='#d73027', lw=1.5),
                    color='#d73027', fontsize=10, fontweight='bold')
    ax.set_title(title, fontweight='bold')
    ax.set_xlabel('Month')
    ax.set_ylabel('Level (%)')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.setp(ax.get_xticklabels(), rotation=30, ha='right', fontsize=8)

fig.suptitle('Preattentive Attributes – Four Ways to Highlight the Last 3 Months',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

# 3. Working on Your Dataset

## Dataset: E-commerce Customer Churn Dataset

This section applies the preattentive attribute principles from Chapter 4 (pp. 99–119) to the Churn data. For each visualization:
- **BEFORE**: all visual elements carry equal weight — viewer must scan everything
- **AFTER**: preattentive attributes (color, size, position) direct the eye to the single key insight

**Key insight to communicate:** *Churned customers show significantly more customer service calls and higher cart abandonment rates.*

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

churn = pd.read_csv("ecommerce_customer_churn_dataset.csv")
churn['Age'] = churn['Age'].where(churn['Age'] <= 100)

print(f"Churn dataset shape: {churn.shape}")
print(f"Churn rate: {churn['Churned'].mean()*100:.1f}%")
print(f"\nMean Customer Service Calls — Retained: "
      f"{churn[churn['Churned']==0]['Customer_Service_Calls'].mean():.2f}")
print(f"Mean Customer Service Calls — Churned:  "
      f"{churn[churn['Churned']==1]['Customer_Service_Calls'].mean():.2f}")
```

### Preattentive Attribute 1: Color

**Task:** Show which behavioral metric most distinguishes churned from retained customers.

- **BEFORE**: Bars are all the same color; the viewer cannot immediately see which group differs.
- **AFTER**: Churned customers highlighted in red across all features; retained in grey. The eye goes straight to the red bars.

```python
# Compare mean values of key features between churned and retained customers
features = [
    'Customer_Service_Calls', 'Cart_Abandonment_Rate', 'Days_Since_Last_Purchase',
    'Login_Frequency', 'Session_Duration_Avg', 'Returns_Rate'
]
feat_labels = [
    'Customer Service Calls', 'Cart Abandonment Rate', 'Days Since Last Purchase',
    'Login Frequency', 'Session Duration (avg)', 'Returns Rate'
]

retained_means = [churn[churn['Churned']==0][f].mean() for f in features]
churned_means  = [churn[churn['Churned']==1][f].mean() for f in features]

fig, axes = plt.subplots(1, 2, figsize=(15, 5.5), sharey=True)
x = range(len(features))

# BEFORE: same color
w = 0.35
axes[0].bar([i - w/2 for i in x], retained_means, width=w, color='#4472C4', label='Retained')
axes[0].bar([i + w/2 for i in x], churned_means,  width=w, color='#ED7D31', label='Churned')
axes[0].set_xticks(list(x))
axes[0].set_xticklabels(feat_labels, rotation=30, ha='right', fontsize=9)
axes[0].set_title('BEFORE: Two arbitrary colors\n(No visual hierarchy — must read carefully)',
                  fontsize=10, fontweight='bold')
axes[0].legend(frameon=False)
for spine in ['top', 'right']:
    axes[0].spines[spine].set_visible(False)

# AFTER: retained = grey, churned = accent red
axes[1].bar([i - w/2 for i in x], retained_means, width=w, color='#cccccc', label='Retained')
axes[1].bar([i + w/2 for i in x], churned_means,  width=w, color='#d73027', label='Churned (accent)')
axes[1].set_xticks(list(x))
axes[1].set_xticklabels(feat_labels, rotation=30, ha='right', fontsize=9)
axes[1].set_title('AFTER: Grey = retained, Red = churned\n“Churned customers call support more and abandon more carts”',
                  fontsize=10, fontweight='bold')
axes[1].legend(frameon=False)
for spine in ['top', 'right']:
    axes[1].spines[spine].set_visible(False)

fig.suptitle('Churn Dataset – Preattentive Color: Feature Means by Churn Status',
             fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

### Preattentive Attribute 2: Size

Displaying key statistics as large numbers lets the viewer absorb the most important facts in under a second. Size pre-attentively signals “this is important.”

```python
overall_churn = churn['Churned'].mean() * 100
high_calls_churn = churn[churn['Customer_Service_Calls'] > 5]['Churned'].mean() * 100
high_cart_churn  = churn[churn['Cart_Abandonment_Rate'] > 0.7]['Churned'].mean() * 100

fig, axes = plt.subplots(1, 3, figsize=(13, 4))

for ax, pct, label, color, note in [
    (axes[0], overall_churn, 'Overall churn rate', '#888888', 'Baseline'),
    (axes[1], high_calls_churn, 'Churn rate when\ncustomer calls > 5', '#d73027',
     'Key risk signal'),
    (axes[2], high_cart_churn,  'Churn rate when\ncart abandonment > 70%', '#e67e22',
     'Secondary risk signal'),
]:
    ax.text(0.5, 0.60, f'{pct:.0f}%', transform=ax.transAxes,
            fontsize=46, fontweight='bold', color=color, ha='center', va='center')
    ax.text(0.5, 0.25, label, transform=ax.transAxes,
            fontsize=11, color='#555555', ha='center', va='center')
    ax.text(0.5, 0.05, note, transform=ax.transAxes,
            fontsize=8, color='#aaaaaa', ha='center', va='bottom', style='italic')
    ax.axis('off')
    ax.set_facecolor('#fafafa')
    rect = plt.Rectangle((0.03, 0.03), 0.94, 0.94, fill=False,
                          edgecolor=color, lw=2, transform=ax.transAxes)
    ax.add_patch(rect)

fig.suptitle('Churn Dataset – Preattentive Size: Key Risk Statistics at a Glance',
             fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

### Preattentive Attribute 3: Position + Color

In scatter plots, **position** encodes the data values while **color** separates the groups. Plotting churned customers in a bold accent color on top of retained (grey) customers makes the separation (or lack thereof) immediately visible.

```python
sample = churn.sample(2000, random_state=42)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# BEFORE: both groups same color
axes[0].scatter(sample['Customer_Service_Calls'], sample['Cart_Abandonment_Rate'],
                color='#4472C4', alpha=0.25, s=12)
axes[0].set_xlabel('Customer Service Calls')
axes[0].set_ylabel('Cart Abandonment Rate')
axes[0].set_title('BEFORE: All same color\n(Cannot see which group is churned)',
                  fontsize=10, fontweight='bold')
for spine in ['top', 'right']:
    axes[0].spines[spine].set_visible(False)

# AFTER: retained = grey (plotted first), churned = accent red (on top)
retained_s = sample[sample['Churned'] == 0]
churned_s  = sample[sample['Churned'] == 1]
axes[1].scatter(retained_s['Customer_Service_Calls'], retained_s['Cart_Abandonment_Rate'],
                color='#cccccc', alpha=0.3, s=12, label='Retained', zorder=1)
axes[1].scatter(churned_s['Customer_Service_Calls'], churned_s['Cart_Abandonment_Rate'],
                color='#d73027', alpha=0.5, s=20, label='Churned', zorder=2)
axes[1].set_xlabel('Customer Service Calls')
axes[1].set_ylabel('Cart Abandonment Rate')
axes[1].set_title('AFTER: Churned in red (accent), retained in grey\n“Churned customers cluster toward high calls + high abandonment”',
                  fontsize=10, fontweight='bold')
axes[1].legend(frameon=False, fontsize=9)
for spine in ['top', 'right']:
    axes[1].spines[spine].set_visible(False)

fig.suptitle('Churn Dataset – Preattentive Position + Color: Scatter Plot',
             fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

### Conclusions

**What the Churn data shows (with preattentive attributes applied):**

- **Color insight**: By plotting retained customers in grey and churned in red, the grouped bar comparison immediately shows that churned customers make more service calls and abandon more carts — without requiring the viewer to read every bar.

- **Size insight**: Customers who made more than 5 service calls have a substantially higher churn rate than the baseline. This elevated risk is best communicated as a large bold number that the viewer absorbs instantly.

- **Position + Color insight**: The scatter plot of Customer Service Calls vs. Cart Abandonment Rate reveals that churned customers (red dots) tend to cluster in the upper-right quadrant — they both call more AND abandon more. Position encodes the data; color separates the groups.

**Preattentive design decisions:**
1. Always plot the grey (non-highlight) group first, so the accent-color group renders on top.
2. Use a muted grey (`#cccccc`) for the background group, not a competing color.
3. State the insight in the chart title, not just a neutral description.
4. Avoid more than one accent color per chart — two accents split the viewer’s attention.

# 4. Storytelling With Data Plot

Reproduce a graph from pp. 99–119 of *Storytelling With Data* (Ch. 4, first half — Focus your audience’s attention).

**Choice:** The **“grey all, one accent color” horizontal bar chart** — the signature example of this chapter.
- All non-highlight bars in neutral grey
- The bar carrying the key message in a bold accent color
- Chart title states the insight directly

**Data:** Churn dataset — churn rate (%) by top-8 countries. The country with the highest churn is highlighted.

```python
# Storytelling With Data pp. 99-119
# Grey all elements; use ONE accent color on the key insight.
# Data: Churn rate (%) by top-8 countries

import matplotlib.pyplot as plt

top_countries = churn['Country'].value_counts().head(8).index
churn_by_country = (
    churn[churn['Country'].isin(top_countries)]
    .groupby('Country')['Churned']
    .mean() * 100
).sort_values()

max_country = churn_by_country.idxmax()
colors = ['#d73027' if c == max_country else '#cccccc'
          for c in churn_by_country.index]

fig, ax = plt.subplots(figsize=(9, 4.5))
bars = ax.barh(churn_by_country.index, churn_by_country.values,
               color=colors, height=0.6)

# Data labels
for bar, (country, val) in zip(bars, churn_by_country.items()):
    is_max = country == max_country
    ax.text(val + 0.2, bar.get_y() + bar.get_height() / 2,
            f'{val:.1f}%', va='center', fontsize=10,
            fontweight='bold' if is_max else 'normal',
            color='#d73027' if is_max else '#888888')

# Add overall average reference line
overall = churn['Churned'].mean() * 100
ax.axvline(overall, color='#555555', linestyle='--', lw=1.2)
ax.text(overall + 0.2, -0.7, f'Overall avg:\n{overall:.1f}%',
        color='#555555', fontsize=8, va='top')

# Remove all chart junk
for spine in ax.spines.values():
    spine.set_visible(False)
ax.xaxis.set_visible(False)
ax.set_xlim(0, 40)

# Insight-stating title at top-left
ax.set_title(
    f'{max_country} has the highest customer churn rate\n'
    '% of customers who churned, by top-8 country markets',
    fontsize=11, fontweight='bold', loc='left', pad=14
)

plt.tight_layout()
plt.show()
```
